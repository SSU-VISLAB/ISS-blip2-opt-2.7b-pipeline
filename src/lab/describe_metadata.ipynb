```python
from transformers import Blip2ForConditionalGeneration
```

```python
def load_model():
    return Blip2ForConditionalGeneration.from_pretrained(
        "Salesforce/blip2-opt-2.7b",
    )  # 
```

```python
model = load_model()
model = model.cuda()
print(model)
```
